```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from typing import List, Dict, Optional, Tuple

# 设置学术图表样式
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'pdf.fonttype': 42,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.figsize': (8, 4)
})

def plot_individual_metric_comparison(
    data1_list: List[Dict[str, float]],
    data2_list: List[Dict[str, float]],
    output_dir: str,
    metric_key: str,
    metric_name: str,
    group1_name: str = "Group1",
    group2_name: str = "Group2"
) -> None:
    """生成单个指标的对比图表"""
    group1_values = [d[metric_key] for d in data1_list if not np.isnan(d[metric_key])]
    group2_values = [d[metric_key] for d in data2_list if not np.isnan(d[metric_key])]

    x_labels = [os.path.splitext(d['filename'])[0] for d in data1_list]
    x = np.arange(len(x_labels))
    bar_width = 0.35

    fig, ax = plt.subplots(figsize=(6, 6))
    bars1 = ax.bar(x - bar_width/2, group1_values, bar_width,
                  color='#219ebc', alpha=0.8, edgecolor='black', label=group1_name)
    bars2 = ax.bar(x + bar_width/2, group2_values, bar_width,
                  color='#023047', alpha=0.8, edgecolor='black', label=group2_name)
    # bars1 = ax.bar(x - bar_width/2, group1_values, bar_width,
    #               color='#6CA2E7', alpha=0.8, edgecolor='black', label=group1_name)
    # bars2 = ax.bar(x + bar_width/2, group2_values, bar_width,
    #               color='#00B4D9', alpha=0.8, edgecolor='black', label=group2_name)
    # 添加数值标签
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height*100:.1f}%',
                        xy=(bar.get_x() + bar.get_width()/2, height),
                        xytext=(0, 3), textcoords="offset points",
                        ha='center', va='bottom', fontsize=8)

    # 设置图表属性
    # ax.set_xlabel('Sample')
    ax.set_ylabel(metric_name)
    ax.set_title(f'{metric_name} Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels, rotation=45, ha='right')
    ax.set_ylim(0, max(max(group1_values), max(group2_values))*1.2)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.legend(loc='upper right')

    # 保存图表
    filename = f"{metric_name.lower().replace(' ', '_')}_comparison.png"
    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, filename), dpi=300, bbox_inches='tight')
    plt.close()

def generate_comparison_from_csv(
    input_csv: str,
    output_dir: str,
    group1_name: str = "Baseline",
    group2_name: str = "AI_Assisted"
) -> None:
    """从CSV生成对比图表"""
    # 读取CSV文件
    df = pd.read_csv(input_csv)

    # 验证必要列是否存在
    required_columns = [
        'Filename',
        f'{group1_name}_Acc', f'{group1_name}_AUC', f'{group1_name}_F1',
        f'{group2_name}_Acc', f'{group2_name}_AUC', f'{group2_name}_F1'
    ]
    if not all(col in df.columns for col in required_columns):
        missing = [col for col in required_columns if col not in df.columns]
        raise ValueError(f"Missing required columns: {missing}")

    # 转换为数据列表
    data1_list = []
    data2_list = []

    for _, row in df.iterrows():
        data1 = {
            'filename': row['Filename'],
            'mean_acc': row[f'{group1_name}_Acc'],
            'auc': row[f'{group1_name}_AUC'],
            'f1': row[f'{group1_name}_F1']
        }
        data2 = {
            'filename': row['Filename'],
            'mean_acc': row[f'{group2_name}_Acc'],
            'auc': row[f'{group2_name}_AUC'],
            'f1': row[f'{group2_name}_F1']
        }
        data1_list.append(data1)
        data2_list.append(data2)

    # 创建输出目录
    os.makedirs(output_dir, exist_ok=True)

    # 生成图表
    plot_individual_metric_comparison(data1_list, data2_list, output_dir,
                                     'mean_acc', 'Accuracy', group1_name, group2_name)
    plot_individual_metric_comparison(data1_list, data2_list, output_dir,
                                     'auc', 'AUC', group1_name, group2_name)
    plot_individual_metric_comparison(data1_list, data2_list, output_dir,
                                     'f1', 'F1 Score', group1_name, group2_name)

    print(f"\nCharts saved to: {output_dir}")

if __name__ == "__main__":
    # 配置参数（只需修改以下路径）
    input_csv = "/content/drive/MyDrive/AI_assistant_result/per_sample_comparison.csv"  # 输入的CSV文件路径
    output_dir = "/content/drive/MyDrive/AI_assistant_result/figure-qwen3-c3"             # 输出目录

    # 生成图表（保持组名与CSV列名一致）
    generate_comparison_from_csv(
        input_csv=input_csv,
        output_dir=output_dir,
        group1_name="Baseline",      # 与CSV列名前缀一致
        group2_name="AI_Assisted"    # 与CSV列名前缀一致
    )
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from typing import List, Dict, Optional, Tuple, Any

# Set academic figure style
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'pdf.fonttype': 42, # Use TrueType fonts for better compatibility
    'ps.fonttype': 42,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    # 'figure.figsize': (8, 6), # Set figure size within the plot function
    'figure.dpi': 300 # Set default dpi
})

def plot_individual_metric_comparison(
    group1_values: np.ndarray,
    group2_values: np.ndarray,
    x_labels: List[str],
    output_dir: str,
    metric_key: str, # Used for filename
    metric_name: str, # Used for Y-axis label and title
    group1_name: str = "Group1",
    group2_name: str = "Group2"
) -> None:
    """
    Generates a comparison bar chart for a single metric between two groups,
    matching the style of the provided image example.

    Args:
        group1_values (np.ndarray): Array of metric values for group 1.
        group2_values (np.ndarray): Array of metric values for group 2.
        x_labels (List[str]): Labels for the x-axis (e.g., sample names).
        output_dir (str): Directory to save the output figure.
        metric_key (str): Key string for the metric (e.g., 'Acc', 'AUC', 'F1')
                          used for generating the output filename.
        metric_name (str): Display name of the metric (e.g., 'Accuracy', 'AUC')
                           used for the Y-axis label and plot title.
        group1_name (str): Display name for group 1 (for legend).
        group2_name (str): Display name for group 2 (for legend).
    """
    # Filter out samples where the metric is NaN for either group
    # Ensure x_labels align with non-NaN values
    valid_mask = ~np.isnan(group1_values) & ~np.isnan(group2_values)
    filtered_group1_values = group1_values[valid_mask]
    filtered_group2_values = group2_values[valid_mask]
    filtered_x_labels = [x_labels[i] for i, valid in enumerate(valid_mask) if valid]

    if not filtered_x_labels:
        print(f"Warning: No valid data points found for {metric_name} comparison.")
        return

    x = np.arange(len(filtered_x_labels))
    bar_width = 0.35

    # Dynamic figure width based on number of samples, aiming for a reasonable aspect ratio
    # Adjust these multipliers/base sizes to fine-tune the chart's width and height
    fig_width_per_sample = 0.8 # inches per sample pair
    base_fig_width = 2 # inches for margins etc.
    fig_width = max(base_fig_width, len(filtered_x_labels) * fig_width_per_sample)
    fig_height = 6 # Keep height fixed or adjust relative to width

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))

    # Use colors similar to the provided image example
    color_baseline = '#219ebc' # Steel blue-like
    color_ai_assisted = '#023047' # Orange-like
    # color_baseline = '#559DC7' # Steel blue-like
    # color_ai_assisted = '#FF993E' # Orange-like

    bars1 = ax.bar(x - bar_width/2, filtered_group1_values, bar_width,
                   color=color_baseline, alpha=0.9, edgecolor='black', label=group1_name) # Increased alpha slightly
    bars2 = ax.bar(x + bar_width/2, filtered_group2_values, bar_width,
                   color=color_ai_assisted, alpha=0.9, edgecolor='black', label=group2_name) # Increased alpha slightly

    # --- Re-introduced Bar value annotations as percentages, matching image style ---
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            # Only annotate if height is not NaN and above a minimal threshold
            if not np.isnan(height) and height >= 0.005: # Add a small threshold to avoid clutter on tiny bars
                 # Format as Percentage with 1 decimal place, matching image
                 ax.annotate(f'{height*100:.1f}%',
                            xy=(bar.get_x() + bar.get_width()/2, height),
                            xytext=(0, 3), textcoords="offset points",
                            ha='center', va='bottom', fontsize=9) # Font size similar to image

    # Set chart properties
    # --- Set Title as requested, matching image format ---
    # ax.set_title(f'{metric_name} Comparison by Sample')

    # --- Set X-axis Label as in the image ---
    # ax.set_xlabel('Sample')
    ax.set_ylabel(metric_name) # Y-axis label shows metric name

    ax.set_xticks(x)
    # --- Always slant X-axis labels as requested and shown in image ---
    rotation = 45
    ax.set_xticklabels(filtered_x_labels, rotation=rotation, ha='right')

    # Determine y-axis limits. Ensure 0 is included and add sufficient padding for labels.
    max_value = max(filtered_group1_values.max() if filtered_group1_values.size else 0,
                    filtered_group2_values.max() if filtered_group2_values.size else 0)
    # Add more padding at the top to ensure percentage labels fit comfortably above bars
    ax.set_ylim(0, max(1.0, max_value * 1.25)) # Ensure Y-axis goes at least to 1.0, add 25% padding or more if needed
    # Set y-ticks to match the image style (e.g., 0.0, 0.2, 0.4...)
    ax.set_yticks(np.arange(0, ax.get_ylim()[1] + 0.01, 0.2)) # Ticks every 0.2, up to max_value or 1.0

    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # --- Legend placement matching image (inside, upper left) ---
    ax.legend(loc='upper left') # Frame is on by default

    # Ensure spines (axes lines) are visible
    for spine in ax.spines.values():
        spine.set_visible(True)


    # Save chart
    # Ensure metric_key is suitable for filename (lowercase, no spaces)
    safe_metric_key = metric_key.lower().replace(' ', '_').replace('.', '')
    filename = f"{safe_metric_key}_comparison.png"
    filepath = os.path.join(output_dir, filename)

    # Use tight_layout without rect as legend is inside
    plt.tight_layout()

    try:
        plt.savefig(filepath, dpi=300, bbox_inches='tight') # bbox_inches='tight' can help with rotated labels
        print(f"Saved chart: {filepath}")
    except Exception as e:
        print(f"Error saving chart {filepath}: {e}")

    plt.close(fig) # Close the figure to free memory


def generate_comparison_from_csv(
    input_csv: str,
    output_dir: str,
    group1_name: str = "Baseline",
    group2_name: str = "AI_Assisted"
) -> None:
    """
    Generates comparison charts for Accuracy, AUC, and F1 Score from a CSV,
    matching the style of the provided image example.

    Args:
        input_csv (str): Path to the input CSV file.
        output_dir (str): Directory to save the output figures.
        group1_name (str): Name of the first group (matches CSV column prefix).
        group2_name (str): Name of the second group (matches CSV column prefix).
    """
    # Read CSV file
    try:
        df = pd.read_csv(input_csv)
    except FileNotFoundError:
        print(f"Error: Input CSV not found at {input_csv}")
        return
    except pd.errors.EmptyDataError:
        print(f"Error: Input CSV is empty {input_csv}")
        return
    except Exception as e:
        print(f"Error reading CSV {input_csv}: {e}")
        return

    # Define metrics and their display names.
    # Note: The image only shows Accuracy. This code generates for Acc, AUC, F1
    # but applies the image's style to all.
    metrics_to_plot = {
        'Acc': 'Accuracy',
        'AUC': 'AUC',
        'F1': 'F1 Score'
    }

    # Validate necessary columns exist using the defined metrics
    required_columns = ['Filename']
    for metric_key in metrics_to_plot.keys():
        required_columns.append(f'{group1_name}_{metric_key}')
        required_columns.append(f'{group2_name}_{metric_key}')

    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Error: Missing required columns in CSV: {missing_columns}")
        return

    # Extract filenames for x-axis labels
    # Use apply and splitext for robustness, handle potential non-string filenames
    try:
        x_labels_all = df['Filename'].apply(lambda x: os.path.splitext(str(x))[0]).tolist()
    except KeyError:
         print("Error: 'Filename' column not found in CSV.")
         return
    except Exception as e:
         print(f"Error processing 'Filename' column: {e}")
         return


    # Create output directory
    os.makedirs(output_dir, exist_ok=True)

    # Generate chart for each metric
    for metric_key, metric_name in metrics_to_plot.items():
        col1 = f'{group1_name}_{metric_key}'
        col2 = f'{group2_name}_{metric_key}'

        # Extract columns as numpy arrays
        group1_values_all = df[col1].values
        group2_values_all = df[col2].values

        print(f"\nGenerating chart for {metric_name}...")

        plot_individual_metric_comparison(
            group1_values=group1_values_all,
            group2_values=group2_values_all,
            x_labels=x_labels_all,
            output_dir=output_dir,
            metric_key=metric_key,
            metric_name=metric_name,
            group1_name=group1_name,
            group2_name=group2_name
        )

    print(f"\nChart generation process completed. Check directory: {output_dir}")


if __name__ == "__main__":
    # Configuration parameters
    # Ensure these paths are correct for your environment
    input_csv_path = "/content/drive/MyDrive/AI_assistant_result/per_sample_comparison.csv"  # Input CSV file path
    output_directory = "/content/drive/MyDrive/AI_assistant_result/figure-qwen3-c3"         # Output directory

    # Generate charts (ensure group names match CSV column prefixes)
    generate_comparison_from_csv(
        input_csv=input_csv_path,
        output_dir=output_directory,
        group1_name="Baseline",      # Matches prefix like "Baseline_Acc"
        group2_name="AI_Assisted"    # Matches prefix like "AI_Assisted_Acc"
    )
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from typing import List, Dict, Optional, Tuple, Any

# Set academic figure style
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'pdf.fonttype': 42, # Use TrueType fonts for better compatibility
    'ps.fonttype': 42,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.dpi': 300 # Set default dpi
})

def plot_individual_metric_comparison(
    group1_values: np.ndarray,
    group2_values: np.ndarray,
    x_labels: List[str],
    output_dir: str,
    metric_key: str, # Used for filename
    metric_name: str, # Used for Y-axis label and title
    group1_name: str = "Group1",
    group2_name: str = "Group2",
    pair_color: str = '#4682B4', # Main color for this metric pair
    hatch_pattern1: Optional[str] = None # Hatch pattern for Group 1
) -> None:
    """
    Generates a comparison bar chart for a single metric between two groups,
    with customizable colors and patterns per metric, showing white background
    for hatched areas.

    Args:
        group1_values (np.ndarray): Array of metric values for group 1.
        group2_values (np.ndarray): Array of metric values for group 2.
        x_labels (List[str]): Labels for the x-axis (e.g., sample names).
        output_dir (str): Directory to save the output figure.
        metric_key (str): Key string for the metric (e.g., 'Acc', 'AUC', 'F1')
                          used for generating the output filename.
        metric_name (str): Display name of the metric (e.g., 'Accuracy', 'AUC')
                           used for the Y-axis label and plot title.
        group1_name (str): Display name for group 1 (for legend).
        group2_name (str): Display name for group 2 (for legend).
        pair_color (str): The main color used for this metric pair (hatch lines for Group1, solid fill for Group2).
        hatch_pattern1 (Optional[str]): Hatch pattern string for Group 1 bars.
    """
    # Filter out samples where the metric is NaN for either group
    valid_mask = ~np.isnan(group1_values) & ~np.isnan(group2_values)
    filtered_group1_values = group1_values[valid_mask]
    filtered_group2_values = group2_values[valid_mask]
    filtered_x_labels = [x_labels[i] for i, valid in enumerate(valid_mask) if valid]

    if not filtered_x_labels:
        print(f"Warning: No valid data points found for {metric_name} comparison.")
        return

    x = np.arange(len(filtered_x_labels))
    bar_width = 0.35

    # Dynamic figure width based on number of samples
    fig_width_per_sample = 0.8
    base_fig_width = 2
    fig_width = max(base_fig_width, len(filtered_x_labels) * fig_width_per_sample)
    fig_height = 6

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))

    # Plot bars: Group1 (Baseline) with white fill and colored hatch, Group2 (AI_Assisted) with solid color
    bars1 = ax.bar(x - bar_width/2, filtered_group1_values, bar_width,
                   color='white', # White fill for hatched bars
                   edgecolor=pair_color, # Color of hatch lines and bar outline
                   hatch=hatch_pattern1,
                   label=group1_name)
    bars2 = ax.bar(x + bar_width/2, filtered_group2_values, bar_width,
                   color=pair_color, # Solid color fill for Group2
                   edgecolor='black', # Outline color for solid bars
                   hatch=None, # No hatch for Group2
                   label=group2_name)

    # Add Bar value annotations as percentages
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            if not np.isnan(height) and height >= 0.005:
                 ax.annotate(f'{height*100:.1f}%',
                            xy=(bar.get_x() + bar.get_width()/2, height),
                            xytext=(0, 3), textcoords="offset points",
                            ha='center', va='bottom', fontsize=9)

    # Set chart properties
    # ax.set_title(f'{metric_name} Comparison by Sample')
    # ax.set_xlabel('Sample')
    ax.set_ylabel(metric_name)

    ax.set_xticks(x)
    # Always slant X-axis labels
    rotation = 45
    ax.set_xticklabels(filtered_x_labels, rotation=rotation, ha='right')

    # Determine y-axis limits and ticks
    max_value = max(filtered_group1_values.max() if filtered_group1_values.size else 0,
                    filtered_group2_values.max() if filtered_group2_values.size else 0)
    y_upper_limit = max(1.05, max_value * 1.25) # Ensure space above max value and 1.0
    ax.set_ylim(0, y_upper_limit)
    ax.set_yticks(np.arange(0, y_upper_limit + 0.01, 0.2)) # Ticks every 0.2

    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Legend placement matching image (inside, upper left)
    ax.legend(loc='upper left')

    # Ensure spines are visible
    for spine in ax.spines.values():
        spine.set_visible(True)

    # Save chart
    safe_metric_key = metric_key.lower().replace(' ', '_').replace('.', '')
    filename = f"{safe_metric_key}_comparison.png"
    filepath = os.path.join(output_dir, filename)

    plt.tight_layout()

    try:
        plt.savefig(filepath, dpi=300, bbox_inches='tight')
        print(f"Saved chart: {filepath}")
    except Exception as e:
        print(f"Error saving chart {filepath}: {e}")

    plt.close(fig) # Close the figure to free memory


def generate_comparison_from_csv(
    input_csv: str,
    output_dir: str,
    group1_name: str = "Baseline",
    group2_name: str = "AI_Assisted"
) -> None:
    """
    Generates comparison charts for Accuracy, AUC, and F1 Score from a CSV,
    applying different color and pattern styles per metric with white hatched areas.

    Args:
        input_csv (str): Path to the input CSV file.
        output_dir (str): Directory to save the output figures.
        group1_name (str): Name of the first group (matches CSV column prefix).
        group2_name (str): Name of the second group (matches CSV column prefix).
    """
    # Read CSV file
    try:
        df = pd.read_csv(input_csv)
    except FileNotFoundError:
        print(f"Error: Input CSV not found at {input_csv}")
        return
    except pd.errors.EmptyDataError:
        print(f"Error: Input CSV is empty {input_csv}")
        return
    except Exception as e:
        print(f"Error reading CSV {input_csv}: {e}")
        return

    # Define metrics, their display names, AND their specific styles (main color and Group1 hatch)
    # Based on the second image example, with white background for hatch.
    metrics_styles = {
        'Acc': {'name': 'Accuracy', 'pair_color': '#4c72b0', 'hatch_pattern1': '///'}, # Blue color, hatched Baseline
        'AUC': {'name': 'AUC', 'pair_color': '#dd8452', 'hatch_pattern1': '///'}, # Orange color, hatched Baseline
        'F1':  {'name': 'F1 Score', 'pair_color': '#55a868', 'hatch_pattern1': '...'}  # Green color, dotted Baseline
    }
    # metrics_styles = {
    #     'Acc': {'name': 'Accuracy', 'pair_color': '#8ecae6', 'hatch_pattern1': '///'}, # Blue color, hatched Baseline
    #     'AUC': {'name': 'AUC', 'pair_color': '#219ebc', 'hatch_pattern1': '///'}, # Orange color, hatched Baseline
    #     'F1':  {'name': 'F1 Score', 'pair_color': '#023047', 'hatch_pattern1': '...'}  # Green color, dotted Baseline
    # }

    # Validate necessary columns exist using the defined metrics
    required_columns = ['Filename']
    for metric_key in metrics_styles.keys():
        required_columns.append(f'{group1_name}_{metric_key}')
        required_columns.append(f'{group2_name}_{metric_key}')

    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Error: Missing required columns in CSV: {missing_columns}")
        return

    # Extract filenames for x-axis labels
    try:
        x_labels_all = df['Filename'].apply(lambda x: os.path.splitext(str(x))[0]).tolist()
    except KeyError:
         print("Error: 'Filename' column not found in CSV.")
         return
    except Exception as e:
         print(f"Error processing 'Filename' column: {e}")
         return


    # Create output directory
    os.makedirs(output_dir, exist_ok=True)

    # Generate chart for each metric using its specific style
    for metric_key, style in metrics_styles.items():
        col1 = f'{group1_name}_{metric_key}'
        col2 = f'{group2_name}_{metric_key}'

        # Extract columns as numpy arrays
        group1_values_all = df[col1].values
        group2_values_all = df[col2].values

        print(f"\nGenerating chart for {style['name']}...")

        plot_individual_metric_comparison(
            group1_values=group1_values_all,
            group2_values=group2_values_all,
            x_labels=x_labels_all,
            output_dir=output_dir,
            metric_key=metric_key,
            metric_name=style['name'],
            group1_name=group1_name,
            group2_name=group2_name,
            pair_color=style['pair_color'], # Pass the main color for the pair
            hatch_pattern1=style['hatch_pattern1'] # Pass the hatch pattern for Group 1
        )

    print(f"\nChart generation process completed. Check directory: {output_dir}")


if __name__ == "__main__":
    # Configuration parameters
    # Ensure these paths are correct for your environment
    input_csv_path = "/content/drive/MyDrive/AI_assistant_result/per_sample_comparison.csv"  # Input CSV file path
    output_directory = "/content/drive/MyDrive/AI_assistant_result/figure-qwen3-c5"         # Output directory

    # Generate charts (ensure group names match CSV column prefixes)
    generate_comparison_from_csv(
        input_csv=input_csv_path,
        output_dir=output_directory,
        group1_name="Baseline",      # Matches prefix like "Baseline_Acc"
        group2_name="AI_Assisted"    # Matches prefix like "AI_Assisted_Acc"
    )
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from typing import List, Dict, Optional, Tuple, Any

# Set academic figure style
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': 'Times New Roman',
    'pdf.fonttype': 42, # Use TrueType fonts for better compatibility
    'ps.fonttype': 42,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.dpi': 300 # Set default dpi
})

def plot_individual_metric_comparison(
    group1_values_all: np.ndarray, # Receive all values before NaN filtering
    group2_values_all: np.ndarray, # Receive all values before NaN filtering
    x_labels_all: List[str],       # Receive all x_labels before NaN filtering
    output_dir: str,
    metric_key: str, # Used for filename
    metric_name: str, # Used for Y-axis label and title
    group1_name: str = "Group1",
    group2_name: str = "Group2",
    pair_color: str = '#4682B4', # Main color for this metric pair
    hatch_pattern1: Optional[str] = None # Hatch pattern for Group 1 (used for samples *other than* the first valid one)
) -> None:
    """
    Generates a comparison bar chart for a single metric between two groups.
    The first VALID sample plotted is shown with solid bars for both groups,
    while other VALID samples use hatch for Group1 and solid for Group2.

    Args:
        group1_values_all (np.ndarray): Array of all metric values for group 1.
        group2_values_all (np.ndarray): Array of all metric values for group 2.
        x_labels_all (List[str]): All labels for the x-axis.
        output_dir (str): Directory to save the output figure.
        metric_key (str): Key string for the metric (e.g., 'Acc', 'AUC', 'F1')
                          used for generating the output filename.
        metric_name (str): Display name of the metric (e.g., 'Accuracy', 'AUC')
                           used for the Y-axis label and plot title.
        group1_name (str): Display name for group 1 (for legend).
        group2_name (str): Display name for group 2 (for legend).
        pair_color (str): The main color used for this metric pair.
        hatch_pattern1 (Optional[str]): Hatch pattern string for Group 1 bars
                                        for samples *other than* the first valid one.
    """
    # Filter out samples where the metric is NaN for either group for the CURRENT metric
    valid_mask = ~np.isnan(group1_values_all) & ~np.isnan(group2_values_all)
    filtered_group1_values = group1_values_all[valid_mask]
    filtered_group2_values = group2_values_all[valid_mask]
    filtered_x_labels = [x_labels_all[i] for i, valid in enumerate(valid_mask) if valid]

    # --- Correction: Check filtered_x_labels here after filtering ---
    if not filtered_x_labels:
        print(f"Warning: No valid data points found for {metric_name} comparison after filtering NaNs.")
        return
    # Optional: Add check if the first valid sample is the one you expect (e.g., 'mixfound')
    # if filtered_x_labels[0].lower() != 'mixfound':
    #      print(f"Note: The first VALID sample plotted for {metric_name} is '{filtered_x_labels[0]}', not 'mixfound'. "
    #            "It will receive the special 'first sample' styling.")
    # --- End Correction ---


    x = np.arange(len(filtered_x_labels))
    bar_width = 0.35

    # Dynamic figure width based on number of samples
    fig_width_per_sample = 0.8
    base_fig_width = 2
    fig_width = max(base_fig_width, len(filtered_x_labels) * fig_width_per_sample)
    fig_height = 6

    fig, ax = plt.subplots(figsize=(fig_width, fig_height))

    all_bars1 = []
    all_bars2 = []

    # Iterate through each VALID sample to apply conditional styling
    for i in range(len(filtered_x_labels)):
        val1 = filtered_group1_values[i]
        val2 = filtered_group2_values[i]
        current_x = x[i]

        # Determine style based on whether it's the first VALID sample plotted (index 0 in filtered data)
        is_first_valid_sample = (i == 0)

        # Group 1 (Baseline) style
        current_color1 = pair_color if is_first_valid_sample else 'white' # Solid color for first valid, white fill for others
        current_edgecolor1 = 'black' # Consistent black edge for all G1 bars
        current_hatch1 = None if is_first_valid_sample else hatch_pattern1 # No hatch for first valid, specific hatch for others

        # Group 2 (AI_Assisted) style (always solid color, matching the pair_color)
        current_color2 = pair_color
        current_edgecolor2 = 'black' # Consistent black edge for all G2 bars
        current_hatch2 = None # Always no hatch for G2

        # Add label only to the first bar pair for legend
        label1 = group1_name if i == 0 else "" # Label for the first bar plotted
        label2 = group2_name if i == 0 else "" # Label for the first bar plotted

        # Plot the individual bars for this sample
        bar1 = ax.bar(current_x - bar_width/2, val1, bar_width,
                      color=current_color1, edgecolor=current_edgecolor1, hatch=current_hatch1,
                      alpha=0.9, label=label1)
        bar2 = ax.bar(current_x + bar_width/2, val2, bar_width,
                      color=current_color2, edgecolor=current_edgecolor2, hatch=current_hatch2,
                      alpha=0.9, label=label2)

        # Collect the Rectangle objects for annotation later
        all_bars1.append(bar1[0])
        all_bars2.append(bar2[0])

    # Add Bar value annotations as percentages by iterating over collected bars
    for bar in all_bars1 + all_bars2: # Combine lists to iterate over all bars
        height = bar.get_height()
        if not np.isnan(height) and height >= 0.005:
             ax.annotate(f'{height*100:.1f}%',
                        xy=(bar.get_x() + bar.get_width()/2, height),
                        xytext=(0, 3), textcoords="offset points",
                        ha='center', va='bottom', fontsize=9)

    # Set chart properties
    ax.set_title(f'{metric_name} Comparison by Sample')
    ax.set_xlabel('Sample')
    ax.set_ylabel(metric_name)

    ax.set_xticks(x)
    # Always slant X-axis labels
    rotation = 45
    ax.set_xticklabels(filtered_x_labels, rotation=rotation, ha='right')

    # Determine y-axis limits and ticks
    # Need max value from original filtered arrays
    max_value = max(filtered_group1_values.max() if filtered_group1_values.size else 0,
                    filtered_group2_values.max() if filtered_group2_values.size else 0)
    y_upper_limit = max(1.05, max_value * 1.25)
    ax.set_ylim(0, y_upper_limit)
    ax.set_yticks(np.arange(0, y_upper_limit + 0.01, 0.2))

    ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Legend placement matching image (inside, upper left)
    ax.legend(loc='upper left')

    # Ensure spines are visible
    for spine in ax.spines.values():
        spine.set_visible(True)

    # Save chart
    safe_metric_key = metric_key.lower().replace(' ', '_').replace('.', '')
    filename = f"{safe_metric_key}_comparison.png"
    filepath = os.path.join(output_dir, filename)

    plt.tight_layout()

    try:
        plt.savefig(filepath, dpi=300, bbox_inches='tight')
        print(f"Saved chart: {filepath}")
    except Exception as e:
        print(f"Error saving chart {filepath}: {e}")

    plt.close(fig) # Close the figure to free memory


def generate_comparison_from_csv(
    input_csv: str,
    output_dir: str,
    group1_name: str = "Baseline",
    group2_name: str = "AI_Assisted"
) -> None:
    """
    Generates comparison charts for Accuracy, AUC, and F1 Score from a CSV,
    applying different color and pattern styles per metric, with the first
    valid sample having solid bars for both groups.

    Args:
        input_csv (str): Path to the input CSV file.
        output_dir (str): Directory to save the output figures.
        group1_name (str): Name of the first group (matches CSV column prefix).
        group2_name (str): Name of the second group (matches CSV column prefix).
    """
    # Read CSV file
    try:
        df = pd.read_csv(input_csv)
    except FileNotFoundError:
        print(f"Error: Input CSV not found at {input_csv}")
        return
    except pd.errors.EmptyDataError:
        print(f"Error: Input CSV is empty {input_csv}")
        return
    except Exception as e:
        print(f"Error reading CSV {input_csv}: {e}")
        return

    # Define metrics, their display names, AND their specific styles (main color and Group1 hatch for *other* samples)
    metrics_styles = {
        'Acc': {'name': 'Accuracy', 'pair_color': '#4682B4', 'hatch_pattern1': '///'}, # Blue color, hatched Baseline (others)
        'AUC': {'name': 'AUC', 'pair_color': '#FF8C00', 'hatch_pattern1': '///'}, # Orange color, hatched Baseline (others)
        'F1':  {'name': 'F1 Score', 'pair_color': '#55A868', 'hatch_pattern1': '...'}  # Green color, dotted Baseline (others)
    }

    # Validate necessary columns exist using the defined metrics
    required_columns = ['Filename']
    for metric_key in metrics_styles.keys():
        required_columns.append(f'{group1_name}_{metric_key}')
        required_columns.append(f'{group2_name}_{metric_key}')

    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Error: Missing required columns in CSV: {missing_columns}")
        return

    # Extract ALL filenames for x-axis labels (before any NaN filtering)
    try:
        x_labels_all = df['Filename'].apply(lambda x: os.path.splitext(str(x))[0]).tolist()
    except KeyError:
         print("Error: 'Filename' column not found in CSV.")
         return
    except Exception as e:
         print(f"Error processing 'Filename' column: {e}")
         return


    # Create output directory
    os.makedirs(output_dir, exist_ok=True)

    # Generate chart for each metric using its specific style
    for metric_key, style in metrics_styles.items():
        col1 = f'{group1_name}_{metric_key}'
        col2 = f'{group2_name}_{metric_key}'

        # Extract columns as numpy arrays (pass ALL values to plot function)
        group1_values_all = df[col1].values
        group2_values_all = df[col2].values

        print(f"\nGenerating chart for {style['name']}...")

        plot_individual_metric_comparison(
            group1_values_all=group1_values_all,
            group2_values_all=group2_values_all,
            x_labels_all=x_labels_all, # Pass ALL original labels
            output_dir=output_dir,
            metric_key=metric_key,
            metric_name=style['name'],
            group1_name=group1_name,
            group2_name=group2_name,
            pair_color=style['pair_color'],
            hatch_pattern1=style['hatch_pattern1'] # This hatch is used for others, not the first valid one
        )

    print(f"\nChart generation process completed. Check directory: {output_dir}")


if __name__ == "__main__":
    # Configuration parameters
    # Ensure these paths are correct for your environment
    input_csv_path = "/content/drive/MyDrive/AI_assistant_result/per_sample_comparison.csv"  # Input CSV file path
    output_directory = "/content/drive/MyDrive/AI_assistant_result/figure-qwen3-c6"         # Output directory

    # Generate charts (ensure group names match CSV column prefixes)
    generate_comparison_from_csv(
        input_csv=input_csv_path,
        output_dir=output_directory,
        group1_name="Baseline",      # Matches prefix like "Baseline_Acc"
        group2_name="AI_Assisted"    # Matches prefix like "AI_Assisted_Acc"
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

# 数据准备（替换为真实结果）
age_groups = ['<40', '40-60', '>60']
model_a = np.array([73.2, 77.93, 69.43])
model_b = np.array([57.69, 63.02, 67.24])
x = np.arange(len(age_groups))
width = 0.35

# === Fairness Gap 计算：最大偏离平均值 ===
mean_a = np.mean(model_a)
mean_b = np.mean(model_b)
gap_a = np.max(np.abs(model_a - mean_a))
gap_b = np.max(np.abs(model_b - mean_b))

# === 显著性检验（如无分布数据可视为模拟 p 值）===
t_stat, p_val = ttest_ind(model_a, model_b, equal_var=False)
if p_val < 0.001:
    significance = "***"
elif p_val < 0.01:
    significance = "**"
elif p_val < 0.05:
    significance = "*"
else:
    significance = "n.s."

# === 作图 ===
fig, ax = plt.subplots(figsize=(8, 6))

# 柱状图
bar1 = ax.bar(x - width/2, model_a, width, label='Model A', color='#1f77b4', alpha=0.6)
bar2 = ax.bar(x + width/2, model_b, width, label='Model B', color='#ff7f0e', alpha=0.6)

# 折线图
ax.plot(x - width/2, model_a, marker='o', linestyle='-', color='#1f77b4')
ax.plot(x + width/2, model_b, marker='s', linestyle='--', color='#ff7f0e')

# 柱顶数值标签
for i in range(len(x)):
    ax.text(x[i] - width/2, model_a[i] + 0.8, f'{model_a[i]:.2f}', ha='center', fontsize=9)
    ax.text(x[i] + width/2, model_b[i] + 0.8, f'{model_b[i]:.2f}', ha='center', fontsize=9)

# 添加 gap 注解 (基于偏离平均最大值)
ax.annotate(f'gap = {gap_a:.2f}',
            xy=(np.argmax(np.abs(model_a - mean_a)) - 0.6, mean_a + gap_a + 1.5),
            ha='center', color='#1f77b4', fontsize=10)

ax.annotate(f'gap = {gap_b:.2f} {significance} ↑ larger bias',
            xy=(np.argmax(np.abs(model_b - mean_b)) + 0.6, mean_b + gap_b - 4.5),
            ha='center', color='#ff7f0e', fontsize=10)

# 图表设置
ax.set_ylabel('Performance (%)')
ax.set_title('Model Performance Across Age Groups with Fairness Gaps', fontsize=13)
ax.set_xticks(x)
ax.set_xticklabels(age_groups)
ax.set_ylim(50, 85)
ax.legend(frameon=False)
ax.grid(axis='y', linestyle='--', alpha=0.5)

# 显著性图例说明（手动添加）
ax.text(2.6, 83, f'* p < 0.05', fontsize=9)
ax.text(2.6, 81.5, f'** p < 0.01', fontsize=9)
ax.text(2.6, 80, f'*** p < 0.001', fontsize=9)

plt.tight_layout()
plt.savefig("figure1_fairness_gap_combined.svg", format='svg')
plt.savefig("figure1_fairness_gap_combined.png", dpi=600)
plt.show()
```
